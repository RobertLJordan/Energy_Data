```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```
funding= pd.read_excel('/content/drive/MyDrive/9 to 5 Jobs/REAP/Statewide Energy Data/Community Funding Data/ISER Capital Funding Tracker.xlsx', header=2)
funding
```

###Create pivot tables for differnt 'PROJ_TYPE'


```
construction_funding = funding.loc[funding['PROJ_TYPE'] =='Construction']
construction_funding_pivot=construction_funding.pivot_table(index='LOC_ID', values='Total', aggfunc=['sum', len])
```

```
energy_funding = funding.loc[funding['PROJ_TYPE'] =='Energy']
energy_funding_pivot=energy_funding.pivot_table(index=['LOC_ID', 'PROJ_TYPE'], values='Total', aggfunc=['sum', len])
```

```
other_funding = funding.loc[funding['PROJ_TYPE'].isin(['Transitional', 'Construction', 'Health and Safety', 'Equipment',
                                          'Planning', 'Renewal and Replacement', 'Information Systems','Deferred Maintenance', 'Renovation and Remodeling',
                                          'To be determined', 'Economic Assistance','Water / Sewer / Solid Waste', 'Life / Health / Safety',
                                          'Research / Studies / Planning','Information Technology / Systems / Communication',
                                          'Equipment / Commodities', 'Transportation', 'Development','Parks / Recreation', 'Education', 'Climate Change / Erosion',
                                          'Workforce Dev', 'Law & Justice', 'Info Tech', 'Gasline','Transfer', 'In-State Gasline'])]

other_funding_pivot=other_funding.pivot_table(index=['LOC_ID'], values='Total', aggfunc=['sum', len])
```

```
no_construction= funding.loc[funding['PROJ_TYPE'].isin(['Transitional', 'Energy', 'Health and Safety', 'Equipment',
                                          'Planning', 'Renewal and Replacement', 'Information Systems','Deferred Maintenance', 'Renovation and Remodeling',
                                          'To be determined', 'Economic Assistance','Water / Sewer / Solid Waste', 'Life / Health / Safety',
                                          'Research / Studies / Planning','Information Technology / Systems / Communication',
                                          'Equipment / Commodities', 'Transportation', 'Development','Parks / Recreation', 'Education', 'Climate Change / Erosion',
                                          'Workforce Dev', 'Law & Justice', 'Info Tech', 'Gasline','Transfer', 'In-State Gasline'])]

no_construction_pivot= funding.pivot_table(index=['LOC_ID'], values='Total', aggfunc=['sum', len])
```

```
proj_type_pivot_by_year=funding.pivot_table(index=['LOC_ID','PROJ_TYPE'], columns='Year', values='Total', aggfunc=['sum', len]) 
```

```
proj_type_pivot_all_years=funding.pivot_table(index=['LOC_ID','PROJ_TYPE'], values='Total', aggfunc=['sum', len])
```

```
proj_type_pivot_no_communities= funding.pivot_table(index='PROJ_TYPE', values='Total', aggfunc=['sum', len])
```

```
funding_pivot= funding.pivot_table(index='LOC_ID', columns='Year', values='Total')
```

### write pivot tables to csv and excel for analysis


```
construction_funding.to_csv('All construction Projects.csv')
construction_funding_pivot.to_csv('All construction Pivot.csv')
energy_funding_pivot.to_csv('energy funding pivot.csv')
energy_funding.to_csv('All Energy Projects.csv')
other_funding_pivot.to_csv('other funding pivot.csv')
no_construction_pivot.to_csv('no construction proj type funding.csv')
proj_type_pivot_all_years.to_csv('project types by community.csv')
proj_type_pivot_no_communities.round(2).to_csv('pivot no communities.csv')
funding_pivot.to_excel('full funding pivot table 2000-2020.xlsx')
proj_type_pivot_by_year.to_csv('project type by community by year')
proj_type_pivot_all_years.round(4).to_csv('project types by community for all years.csv')
```
